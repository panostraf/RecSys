```python
import calendar
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
item_embedings = pd.read_csv('item_embeddings.csv')
item_embedings
```

```python
# item_embedings.set_index('item_id',inplace=True)
```

# Plot Scores

```python

sse=[]
scores = []
for i in range(2,20):
    kmeans_args = {"init":"k-means++",
                    "n_init":12,
                    "max_iter":500,
                    "random_state":42}

    kmeans=KMeans(n_clusters=i,**kmeans_args)
    kmeans.fit(item_embedings)
    labels = kmeans.labels_
    sse.append(kmeans.inertia_)
    scores.append(silhouette_score(item_embedings, kmeans.labels_))
```

```python
plt.plot([i for i in range(2,20)], sse)
plt.show()

plt.plot([i for i in range(2,20)], scores)
plt.show()
```

# Clustering

```python
kmeans_args = {"init":"k-means++",
                "n_init":50,
                "max_iter":500,
               'algorithm':'elkan',
                "random_state":42}

kmeans=KMeans(n_clusters=4,**kmeans_args)
kmeans.fit(item_embedings)
labels = kmeans.labels_
# print(len(kmeans.labels_))
# features['cluster'] = kmeans.labels_
# features=features[['item_id','feature_value_id','cluster']]

# clustered_features = pd.concat([features,pd.get_dummies(features['cluster'])],axis=1)

```

```python
item_embedings['cluster'] = labels
item_embedings
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit(item_embedings)
x=pca.transform(item_embedings)
x.shape
x = pd.DataFrame(x).rename(columns={0:'d1',1:'d2'})
x['cluster']=item_embedings.cluster



plt.figure(figsize=(16,4))
ax = plt.axes()
ax.set_facecolor('black')
for cluster in item_embedings.cluster.unique():

    plt.scatter(x[x['cluster']==cluster]['d1'],
                x[x['cluster']==cluster]['d2'],
              label=cluster)

centers = pca.transform(kmeans.cluster_centers_)
# plt.fill(c='black')
plt.scatter(centers[:,0],centers[:,1],marker="+",s=1000,c='white')
plt.title("Plot item vector's clusters using PCA")

plt.legend()
```
