1) clean item_features (reduce number of categories)

2) create vector for each item (one hot encoding)

3) clustering of vectors (create 5 clusters of items)

4) create a vector for each session

5) build classifier to classify sessions to the corresponding cluster (Candidate Generation)

6) More Data cleaning, Discard rare items and keep items closer to cluster center (using item_purchases)

7) Samehow relate session to the most appropriate item from the cluster (Ranking - KNN?)

8) Using all items and sessions measure real accuracy

# Clustering on Item Features

```python
# group by itemid and feature category => one hot encoding on features to create the vector
t = item_features\
.groupby(["item_id","feature_category_id"])\
.count()\
.reset_index()\
.sort_values(by="feature_value_id",ascending=False)


# features = t[t['feature_value_id']==1].drop("feature_value_id",axis=1)
# features = features.drop("feature_value_id",axis=1)
features = t.join(pd.get_dummies(t['feature_category_id']))\
.groupby('item_id').sum().reset_index().drop('feature_category_id',axis=1)
features.head()
```

```python
features_ = features[features.columns[2:]]
features_
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features_)
kmeans_args = {"init":"k-means++",
                "n_init":10,
                "max_iter":100,
                "random_state":42}
sse=[]
scores = []
test_n_clusters = 20
for i in range(2,test_n_clusters):
    if i % 5 ==0:
        print(i)
    kmeans=KMeans(n_clusters=i,**kmeans_args)
    kmeans.fit(scaled_features)
    sse.append(kmeans.inertia_)
    scores.append(silhouette_score(scaled_features, kmeans.labels_))
#     print(i)
    

```

```python
plt.figure(figsize=(12,6))
plt.plot([str(i) for i in range(2,test_n_clusters)],sse,label='score')
plt.ylabel("sse")
plt.xlabel("number of clusters")
plt.title("Inertia")
plt.scatter("5",170000,label='Elbow',color='red')
plt.scatter("8",122000,label='Elbow2',color='red')
plt.grid()
plt.show()


plt.figure(figsize=(12,6))
plt.plot([str(i) for i in range(2,test_n_clusters)],scores,label='score')
plt.ylabel("silhouette_score")
plt.xlabel("number of clusters")
plt.title("Silhouette")
plt.scatter("5",0.454,label='Elbow',color='red')
plt.scatter("8",0.392,label='Elbow2',color='red')
plt.legend()
plt.grid()
plt.show()
```

```python
kmeans_args = {"init":"k-means++",
                "n_init":10,
                "max_iter":100,
                "random_state":42}

kmeans=KMeans(n_clusters=12,**kmeans_args)
kmeans.fit(scaled_features)

print(len(kmeans.labels_))
features['cluster'] = kmeans.labels_
features=features[['item_id','feature_value_id','cluster']]

clustered_features = pd.concat([features,pd.get_dummies(features['cluster'])],axis=1)

```

```python
plt.figure(figsize=(16,4))
clustered_features.groupby('cluster').count()['item_id'].plot.bar()
plt.title("Cluster Distribution")
plt.ylabel("Counts")
plt.xlabel("Cluster")

plt.show()

from sklearn.decomposition import PCA

pca=PCA(n_components=2)
pca.fit(scaled_features)
x=pca.transform(scaled_features)
x.shape
x = pd.DataFrame(x).rename(columns={0:'d1',1:'d2'})
x['cluster']=features.cluster

plt.figure(figsize=(16,4))
ax = plt.axes()
ax.set_facecolor('black')
for cluster in features.cluster.unique():

    plt.scatter(x[x['cluster']==cluster]['d1'],
                x[x['cluster']==cluster]['d2'],
              label=cluster)

centers = pca.transform(kmeans.cluster_centers_)
# plt.fill(c='black')
plt.scatter(centers[:,0],centers[:,1],marker="+",s=1000,c='white')
plt.title("Plot item vector's clusters using PCA")

plt.legend()
```
