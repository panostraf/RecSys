```python
import calendar
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.decomposition import PCA
import math
from sklearn.preprocessing import StandardScaler
```

```python
cand_items = pd.read_csv('dressipi_recsys2022/candidate_items.csv', sep=',')
item_features = pd.read_csv('clustered_features.csv', sep=',')
train_purchases = pd.read_csv('train_purchases_cleanned.csv', sep=',')
train_sessions = pd.read_csv('train_sessions_cleanned.csv', sep=',')


```

```python
train_sessions = train_sessions.merge(item_features)
train_sessions
```

```python
df = train_sessions.drop(['cluster','date','feature_value_id'],axis=1)
df = df.groupby('session_id').sum()
```

```python
df = df.merge(item_features[['item_id','cluster']],on='item_id').drop('item_id',axis=1)
df
```

```python
# scaler = StandardScaler()
# scaler.fit(df.iloc[:,:-1] )
# scaler.fit()
# df.iloc[:,:-1] = scaler.transform(df.iloc[:,:-1])
import pickle
# pickle.dump(scaler)
```

```python
df = df.sample(frac=1)
```

```python
samples = df.groupby('cluster').count()['0'].min()
```

```python
d = pd.DataFrame()
for c in df.cluster.unique():
    d = pd.concat([d,df[df.cluster==c].sample(n=samples)])
    
d
```

```python
d.groupby('cluster').count()
```

```python
from sklearn.model_selection import train_test_split
# split = round(int(len(df)*0.7))
# train = df[:split]
# test = df[split+1:]


# x_train = train.iloc[:,:-1]
# y_train = train.iloc[:,-1:]
# x_test = test.iloc[:,:-1]
# y_test = test.iloc[:,-1:]
x_train, x_test, y_train, y_test = train_test_split(
    df.iloc[:,:-1], df.iloc[:,-1:], test_size=0.33, random_state=42)
```

```python
rfclf = RandomForestClassifier( max_depth=None,random_state=0,n_estimators=1000,n_jobs=-1)
rfclf.fit(x_train, y_train)
```

```python
rfclf.score(x_train,y_train)
```

```python
rfclf.score(x_test,y_test)
```

```python
rfclf.feature_importances_
```

```python
rfclf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix

y_pred = rfclf.predict(x_train)
y_true = y_train
cf = confusion_matrix(y_true, y_pred, labels=None, sample_weight=None, normalize=None)
```
