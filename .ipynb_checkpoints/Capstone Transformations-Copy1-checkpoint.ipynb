## Map timestamp and hours to working hours or no working hours


```python
original_data['datetime(UTC)'] = pd.to_datetime(original_data['date'], utc=True)
original_data['datetime'] = original_data['datetime(UTC)'].astype(str)
original_data['time'] = original_data['datetime'].str[11:-6]
original_data['date'] = original_data['datetime'].str[:10]
original_data[['hour', 'min', 'sec']] = original_data['time'].str.split(':', expand=True)
print('before for loop')
original_data['hour'] = original_data['hour'].astype(int)

original_data['hour'] = original_data['hour'].replace([0, 1, 2, 3, 4, 5, 6, 7, 17, 18, 19, 20, 21, 22, 23], 1)
original_data['hour'] = original_data['hour'].replace([8, 9, 10, 11, 12, 13, 14, 15, 16], 0)
original_data.rename({'hour': 'time_of_day'}, axis=1, inplace=True)
original_data
```

## Function to find the date from the timestamp

```python
def findday(date):
    born = datetime.datetime.strptime(date, '%Y-%m-%d').weekday()
    return (calendar.day_name[born])
```

## Map days to 0 and 1 for weekend or not

```python
original_data['day'] = original_data['date'].apply(lambda x: findday(x))
original_data['day'] = original_data['day'].replace('Sunday', 1)
original_data['day'] = original_data['day'].replace('Saturday', 1)
original_data['day'] = original_data['day'].replace(r'[^\d]+', 0, regex=True)
```

```python
original_data
```

## Get the session duration for each session

```python
session_duration = original_data.groupby('session_id')['datetime(UTC)'].apply(lambda x: max(x) - min(x))
```

```python
session_duration
```

```python
original_data.head(60)
```

```python
session_steps = original_data.groupby('session_id')['item_id'].count()
session_steps
```

```python
# time of day is the rounded avg cause if he spend more time from 0 category we want 0 or else 1
time_of_day = original_data.groupby('session_id')['time_of_day'].apply(lambda x: round(sum(x) / len(x)))
time_of_day
```

```python
# same thinking as before for the weekend days
weekday = original_data.groupby('session_id')['day'].apply(lambda x: round(sum(x) / len(x)))
weekday
```

```python
# combine all these in a df with renamed columns
horizontal_stack = pd.concat([session_duration.rename('session_duration'), session_steps.rename('session_steps'),
                              weekday.rename('weekday'), time_of_day.rename('time_of_day')], axis=1)

horizontal_stack
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(test)
kmeans_args = {"init":"random",
                "n_init":10,
                "max_iter":30,
                "random_state":42}
sse=[]
scores = []
test_n_clusters = 20
for i in range(2,test_n_clusters):
    kmeans=KMeans(n_clusters=i,**kmeans_args)
    kmeans.fit(scaled_features)
    sse.append(kmeans.inertia_)
#     scores.append(silhouette_score(scaled_features, kmeans.labels_))
#     print(i)
    
```

```python
plt.plot(range(2,test_n_clusters),sse)
plt.ylabel("sse")
plt.xlabel("clusters")
plt.title("Inertia")
```

```python
plt.plot(range(2,test_n_clusters),scores)
plt.xlabel("clusters")
plt.ylabel("silhouette score")
plt.title("Silhouette score")
```
