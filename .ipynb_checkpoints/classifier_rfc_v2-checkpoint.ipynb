## Create a vector for each item

```python
# group by itemid and feature category => one hot encoding on features to create the vector
t = item_features\
.groupby(["item_id","feature_category_id"])\
.count()\
.reset_index()\
.sort_values(by="feature_value_id",ascending=False)


# features = t[t['feature_value_id']==1].drop("feature_value_id",axis=1)
# features = features.drop("feature_value_id",axis=1)
features = t.join(pd.get_dummies(t['feature_category_id']))\
.groupby('item_id').sum().reset_index().drop('feature_category_id',axis=1)
```

```python
features.head()
```

## Merge vectors of items with train sessions

```python
df =train_sessions.merge(features, how='left', on='item_id')
df.head()
```

```python
# df[df['session_id']==18].iloc[:,5:].T
```

## Create vector for each session
By adding the values of each item together (by session)
.Then convert each value>0 to 1 and the rest to 0

```python
df=df\
.groupby('session_id')\
.sum()\
.reset_index()\
.drop(['item_id'],axis=1)
df.head()
```

```python
sums = []
for col in train_data.columns[:-1]:
#     print(col,":",train_data[col].sum())
    sums.append(train_data[col].sum())
# plt.bar([str(col) for col in train_data.columns[:-1]], sums) 
common = pd.DataFrame([[str(col) for col in train_data.columns[:-1]],sums]).T
common.columns = ['feature_item_id','frequency']
common['z_score'] = common['frequency'].apply(lambda x: (x-common['frequency'].mean())/common['frequency'].std())
```

```python
common['z_score'].hist()
```

```python
features_to_drop = common[common['z_score']<0]['feature_item_id'].tolist()
features_to_drop = [int(i) for i in features_to_drop]
```

## Use only values that are in candidate items & only keep sessions where the y_label appears more than 200 times in the dataset


```python
sample_train_data_items = train_data.groupby("item_id").count()[1].reset_index()
sample_train_data_items.columns=['item_id','purchases']
sample_train_data_items_final = sample_train_data_items[sample_train_data_items['purchases']>500 ]['item_id'].tolist()

train_data = train_data[train_data['item_id'].isin(sample_train_data_items_final)]
train_data = train_data[train_data['item_id'].isin(cand_items['item_id'])]
train_data.drop(features_to_drop,axis=1,inplace=True)
```

# ----PCA----

## Find optimal number of components

```python
max_n_components=20
evrs=[]

for i in range(2,max_n_components+1):
    X=train_data.drop('item_id',axis=1).values
    y=train_data['item_id']

    pca = PCA(n_components=i)
    pca.fit(X)

    X = pca.transform(X)
    # X_test = pca.transform(X_test)
    evr = sum(pca.explained_variance_ratio_)
#     print(evr)
    evrs.append(evr)


```

```python
evr_dict = dict(zip([str(i) for i in range(2,max_n_components+1)],evrs)   )
print(evr_dict)
plt.figure(figsize=(12,6))
plt.plot(evr_dict.keys(),evr_dict.values(),label="Eplained Var Ratio")
plt.grid()
plt.xlabel("Number of components")
plt.ylabel("Explained variance ratio")
plt.vlines(["7"],min(evrs),1,color="orange",label="elbow")
plt.vlines(["13"],min(evrs),1,color="red",label="min loss")
plt.title("PCA Explained Variance Ratio per number of components")

plt.legend()
plt.show()
```
