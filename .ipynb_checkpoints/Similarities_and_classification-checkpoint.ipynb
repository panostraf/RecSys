```python
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt

parameters ={
    'df_path' : 'dressipi_recsys2022/item_features.csv',
    'components_pca_plot' : 50, # How many points the plot will have at the end
    'pca_components': 30,

    
}


item_features = pd.read_csv(parameters['df_path'],sep=',')

item_features.columns = ['item_id','feature_category_id','feature_value_id']
# item_features

value_embeddings = item_features.pivot_table(values='feature_value_id', index='feature_value_id', columns='feature_category_id', aggfunc=lambda x: len(x.unique()))

category_embeddings = item_features.merge(value_embeddings, left_on='feature_value_id' ,right_on=value_embeddings.index).groupby('feature_category_id').sum().drop(['item_id','feature_value_id'],axis=1)
item_embeddings = item_features.merge(category_embeddings,left_on='feature_category_id',right_on=category_embeddings.index).groupby('item_id').sum().drop(['feature_category_id','feature_value_id'],axis=1)



ratios = []

for i in range(3,parameters['components_pca_plot']):
  pca = PCA(n_components=i)
  pca.fit(item_embeddings.values)
  r = pca.explained_variance_ratio_.sum()
  ratios.append(r)

plt.figure(figsize=(12,6))
plt.plot([i for i in range(3,parameters['components_pca_plot'])],ratios)
plt.title('(PCA) Excplained Variance Ratio per Number of Components')
plt.xlabel('Number Of Components')
plt.ylabel('Explained Variance Ratio')
plt.show()


# components = 30
pca = PCA(n_components=parameters['pca_components'])
pca.fit(item_embeddings.values)
reduced_item_embeddings = pca.transform(item_embeddings.values)
# reduced_item_embeddings = item_embeddings


item_embeddings = pd.concat([pd.Series(item_embeddings.index),pd.DataFrame(reduced_item_embeddings)],axis=1)
item_embeddings.to_csv('item_embeddings.csv',index=False)


```

```python
session_embeddings = sessions.copy().merge(item_embeddings).groupby('session_id').sum().drop('item_id',axis=1).reset_index()
purchase_embeddings = purchases.merge(item_embeddings)
# session_embeddings.merge(purchase_embeddings)
purchase_embeddings
```

```python
from sklearn.metrics.pairwise import cosine_similarity


# # (X, Y=None, dense_output=True)[source]¶
# sim.shape
```

```python
import numpy as np
import itertools
```

```python

top1 = 0
top5 = 0
top10 = 0
top20 = 0
top50 = 0
top100 = 0
not_exists = 0

test_iter = 500
for i in range(test_iter):
#     i = random.randint(1,100000)
    sim = cosine_similarity([x_test[i]],x_train)
    
    results = pd.DataFrame(sim.T)


    results = pd.concat([results,pd.Series(y_train)],axis=1) # add item_id (y_train)
    results = pd.concat([results,pd.Series(train_cluster)],axis=1) # add cluster of session
    results = pd.concat([results,train_data.session_id],axis=1) # add session_id
    results.columns = ['similarities','item_id','cluster','session_id'] # rename dataframe
    
    
    results.dropna(axis=0,inplace=True)
    results.sort_values(by='similarities',ascending=False,inplace=True)
    
#     temp = results.reset_index()
    print(results[results.item_id==y_test[i]].reset_index().head())
#     exclude items that belongs to different cluster than the one we believe that the session belongs to
#     results=results[results.cluster==test_cluster[i]]
    
    results = results[results['similarities']>0]
#     if y_test[i] in list(results['item_id'][:100]):
#         top1+=1
    
    results_av_similarity = results.groupby('item_id').mean()['similarities'].sort_values(ascending=False).reset_index()[:100]
    results_av_similarity['item_id'] = results_av_similarity['item_id'].astype(int)
    
#     Get candidates both from av similarity and from cos_sim
    x = results['item_id'][:100].tolist()
    xs = results_av_similarity['item_id'].tolist()
    candidates = list(itertools.chain(*[x,xs]))
    candidates = [int(x) for x in candidates]
#     print("Real Similarity=",results[results['item_id']==y_test[i]]['similarities'])
    
    
    if y_test[i] in candidates:
        
        if y_test[i] == results_av_similarity['item_id'][0]:
            top100 +=1
            top50 +=1
            top20 +=1
            top10 +=1
            top5+=1
            top1+=1
            
        elif y_test[i] in list(results_av_similarity['item_id'])[0:5]:
            top100 +=1
            top50 +=1
            top20 +=1
            top10 +=1
            top5+=1
            
        elif y_test[i] in list(results_av_similarity['item_id'])[0:10]:
            top100 +=1
            top50 +=1
            top20 +=1
            top10 +=1
            
        elif y_test[i] in list(results_av_similarity['item_id'])[0:20]:
            top100 +=1
            top50 +=1
            top20 +=1
            
        elif y_test[i] in list(results_av_similarity['item_id'])[0:50]:
            top100 +=1
            top50 +=1
            
        elif y_test[i] in list(results_av_similarity['item_id'])[0:5]:
            top100 +=1
        else:
            continue
    else:
        if y_test[i] not in y_train:
            not_exists += 1
        continue
            
            
top1 = round((top1/test_iter )*100,3)
top5 = round((top5/test_iter )*100,3)
top10 = round((top10/test_iter )*100,3)
top20 = round((top20/test_iter )*100,3)
top50 = round((top50/test_iter )*100,3)
top100 = round((top100/test_iter )*100,3)
            
#     print(f'----------{i}--------------')
plt.title(f"Frequency of correct answer in experiment of {test_iter} test sessions")
plt.bar(['top1','top5','top10','top20','top50','top100'],[top1,top5,top10,top20,top50,top100])
plt.ylim(0,test_iter)
```
