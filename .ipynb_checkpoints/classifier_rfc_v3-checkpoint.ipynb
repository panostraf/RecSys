Steps:

_____METHOD 1_____
1) Create a vetor for each item in item_features

2) Merge train_sessions with item_features_vectors

3) Group by session to create a vector for this session

4) Add activity measures (such as number of steps per session, duration etc)

5) classifier to learn that this vector sessions matches this item
    
    
or
____METHOD 2______
1) Create a vetor for each item in item_features

2) Merge train_sessions with item_features_vectors

3) Group by session to create a vector for this session

4) Similarity between vector of sessions and vectors of items - along with knn to find the most similar items

```python
cand_items = pd.read_csv('dressipi_recsys2022/candidate_items.csv', sep=',')
item_features = pd.read_csv('dressipi_recsys2022/item_features.csv', sep=',')
train_purchases = pd.read_csv('dressipi_recsys2022/train_purchases.csv', sep=',')
train_sessions = pd.read_csv('dressipi_recsys2022/train_sessions.csv', sep=',')

raw_train_purchases = train_purchases.copy()
raw_train_sessions  = train_sessions.copy()

# check purchases frequency
pf = train_purchases.groupby('item_id').count().reset_index().sort_values(by='item_id')
pf.drop('session_id',axis=1,inplace=True)
pf.columns = ['item_id','frequency']
pf['proba'] = (pf['frequency']/raw_number_of_sessions)*100
proba_mean = pf['proba'].mean()
pf
```

### Remove feature categories that are very rare among all the items, to reduce the one_hot_encoding vector later


```python
threshhold2 = 0.5# original

item_features_frequency = item_features.groupby('feature_category_id').count().reset_index()[['feature_category_id','item_id']]
item_features_frequency.columns = ['feature_category_id','frequency']
frq_mean = item_features_frequency['frequency'].mean()
frq_std = item_features_frequency['frequency'].std()
item_features_frequency['z_score'] = item_features_frequency['frequency'].apply(lambda x: (x-frq_mean)/frq_std)

plt.figure(figsize=(12,6))
plt.scatter(item_features_frequency.z_score,item_features_frequency.feature_category_id)

# plt.hlines(0,0,76,color='blue',label='mean')
# plt.hlines(1,0,76,color='red',label='1std')
# plt.hlines(-1,0,76,color='orange',label='1std')
# # plt.hlines(-0.3,0,76,color='red',label='mean2')
# plt.fill_between((-0.2,75),(2.5,2.5),   color='red',alpha=0.1)#
# plt.fill_between((-0.2,75),(-1,-1),   color='green',alpha=0.1)#

plt.legend(loc='upper right')
plt.title('Frequency of feature category id, using z_score')
plt.show()


item_features_frequency.head()



item_features_frequency = item_features_frequency[item_features_frequency['z_score'] > threshhold2]
item_features_frequency.drop('z_score',axis=1,inplace=True)
frq_mean = item_features_frequency['frequency'].mean()
frq_std = item_features_frequency['frequency'].std()
item_features_frequency['z_score'] = item_features_frequency['frequency'].apply(lambda x: (x-frq_mean)/frq_std)


plt.figure(figsize=(12,6))
plt.scatter(item_features_frequency.feature_category_id,item_features_frequency.z_score)

plt.hlines(0,0,76,color='blue',label='mean')
plt.hlines(1,0,76,color='red',label='1std')
plt.hlines(-1,0,76,color='orange',label='-1std')
plt.fill_between((-0.2,75),(3,3),   color='red',alpha=0.1)#
plt.fill_between((-0.2,75),(-2,-2),   color='green',alpha=0.1)#

plt.legend(loc='upper right')
plt.title('Frequency of feature category id, using z_score')
plt.show()

print("Categories above the specified point were considered and have been removed")
print(f"{len(item_features_frequency)} categories have been kept out of 73")
item_features_frequency.head()
remaining_items = item_features[item_features['feature_category_id'].isin(item_features_frequency['feature_category_id'])]['item_id'].nunique()
print(f"Using the above method from 23691 the remaining unique items are {remaining_items}")
item_features = item_features[item_features['feature_category_id'].isin(item_features_frequency['feature_category_id'])]


```

# ___Data Transformation___

## Create a vector for each item

```python
# group by itemid and feature category => one hot encoding on features to create the vector
t = item_features\
.groupby(["item_id","feature_category_id"])\
.count()\
.reset_index()\
.sort_values(by="feature_value_id",ascending=False)


# features = t[t['feature_value_id']==1].drop("feature_value_id",axis=1)
# features = features.drop("feature_value_id",axis=1)
features = t.join(pd.get_dummies(t['feature_category_id']))\
.groupby('item_id').sum().reset_index().drop('feature_category_id',axis=1)
features.head()
```

# ----PCA----

## Find optimal number of components

```python
max_n_components=20
evrs=[]

for i in range(2,max_n_components+1):
    X=train_data.drop('item_id',axis=1).values
    y=train_data['item_id']

    pca = PCA(n_components=i)
    pca.fit(X)

    X = pca.transform(X)
    # X_test = pca.transform(X_test)
    evr = sum(pca.explained_variance_ratio_)
#     print(evr)
    evrs.append(evr)


```

```python
evr_dict = dict(zip([str(i) for i in range(2,max_n_components+1)],evrs)   )
print(evr_dict)
plt.figure(figsize=(12,6))
plt.plot(evr_dict.keys(),evr_dict.values(),label="Eplained Var Ratio")
plt.grid()
plt.xlabel("Number of components")
plt.ylabel("Explained variance ratio")
plt.vlines(["7"],min(evrs),1,color="orange",label="elbow")
plt.vlines(["13"],min(evrs),1,color="red",label="min loss")
plt.title("PCA Explained Variance Ratio per number of components")

plt.legend()
plt.show()
```
